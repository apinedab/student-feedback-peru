# Student Feedback Dataset (Peru, 2019)
This notebook demonstrates how to use the dataset for basic sentiment analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the dataset (ensure it's in the same directory)
df = pd.read_csv("data/comments_dataset.csv", encoding="ISO-8859-1")
df.head()
```

## Sentiment Distribution

```python
# Count of sentiment labels
sentiment_counts = df['Valor'].value_counts()
print(sentiment_counts)

# Plot sentiment distribution
sentiment_counts.plot(kind='bar', color=['green', 'red'])
plt.title("Distribution of Sentiment Labels")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Word Clouds by Sentiment

```python
# Generate text per sentiment
positive_text = ' '.join(df[df['Valor'].str.lower() == 'positivo']['comentario'].dropna().astype(str))
negative_text = ' '.join(df[df['Valor'].str.lower() == 'negativo']['comentario'].dropna().astype(str))

# Create word clouds
wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
wordcloud_neg = WordCloud(width=800, height=400, background_color='white').generate(negative_text)

# Show word clouds
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis('off')
plt.title("Positive Comments Word Cloud")
plt.show()

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title("Negative Comments Word Cloud")
plt.show()
```
